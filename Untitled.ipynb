```python
import pandas as pd
import numpy as np
import os

# Data klasörünün varlığından emin olalım
os.makedirs('data', exist_ok=True)

# Sentetik SNP verisi oluştur
np.random.seed(42)
n_samples = 150 # Kişi sayısı
snps = ['rs429358', 'rs7412', 'rs6265', 'rs1801133', 'rs4680', 
        'rs1800497', 'rs2228570', 'rs9939609', 'rs1805007', 'rs1333049']

data = {'SampleID': [f'PERSON_{i:03d}' for i in range(n_samples)]}

for snp in snps:
    # Bazı SNP'lerin frekanslarını farklı yapalım
    p_dist = [0.4, 0.4, 0.2] if snp in ['rs429358', 'rs1333049'] else [0.6, 0.3, 0.1]
    data[snp] = np.random.choice([0, 1, 2], size=n_samples, p=p_dist)

# Basit bir hastalık durumu (Risk skoruna göre)
risk_scores = (data['rs429358'] * 0.6 + data['rs1333049'] * 0.4 + np.random.normal(0, 0.1, n_samples))
data['Disease'] = (risk_scores > np.percentile(risk_scores, 70)).astype(int)

df = pd.DataFrame(data)
df.to_csv('data/sample_snp_data.csv', index=False)
print("✅ Veri seti başarıyla oluşturuldu: data/sample_snp_data.csv")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Veriyi oku
df = pd.read_csv('data/sample_snp_data.csv')

# İlk bakış
print(f"Toplam Kişi Sayısı: {len(df)}")
print(f"Hastalık Dağılımı:\n{df['Disease'].value_counts()}")

# SNP'lerin görselleştirmesi
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='rs429358', hue='Disease')
plt.title('Alzheimer ile ilişkili rs429358 SNP Genotip Dağılımı')
plt.xlabel('Genotip (0:AA, 1:AG, 2:GG)')
plt.show()
```

```python
from scipy import stats

# Tüm SNP'leri listele
snp_cols = [c for c in df.columns if c.startswith('rs')]
results = []

for snp in snp_cols:
    # Karşılaştırma tablosu oluştur
    contingency = pd.crosstab(df[snp], df['Disease'])
    
    # Chi-square testi uygula
    chi2, p, dof, ex = stats.chi2_contingency(contingency)
    
    results.append({
        'SNP': snp,
        'P-Value': p,
        'Anlamlı mı?': 'EVET' if p < 0.05 else 'HAYIR'
    })

# Sonuçları tablo olarak göster
analysis_df = pd.DataFrame(results).sort_values('P-Value')
print(analysis_df)
```

```python
# Her SNP için hayali bir risk ağırlığı belirleyelim
weights = {
    'rs429358': 0.5, 
    'rs1333049': 0.4,
    'rs7412': 0.1,
    'rs6265': 0.05
}

# Risk skorunu hesapla
df['PRS'] = sum(df[snp] * weights.get(snp, 0) for snp in snp_cols)

# Risk skorunun dağılımını görelim
plt.figure(figsize=(10, 5))
sns.kdeplot(data=df, x='PRS', hue='Disease', fill=True)
plt.title('Poligenik Risk Skoru (PRS) Dağılımı')
plt.xlabel('Risk Skoru')
plt.ylabel('Yoğunluk')
plt.show()
```

```python
import numpy as np

# P-değerlerini logaritmik ölçeğe çevirelim (-log10 p-value)
analysis_df['-log10_P'] = -np.log10(analysis_df['P-Value'])

# Görselleştirme
plt.figure(figsize=(12, 6))
colors = ['#1f77b4' if p < 1.3 else '#ff7f0e' for p in analysis_df['-log10_P']] # 1.3 yaklaşık p=0.05
plt.scatter(analysis_df['SNP'], analysis_df['-log10_P'], c=colors, s=100)
plt.axhline(y=1.3, color='r', linestyle='--', label='Eşik Değeri (p=0.05)')
plt.title('Genetik İlişki Analizi: Manhattan Plot')
plt.ylabel('-log10(P-Value)')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
import os

# Görseller için klasör oluşturalım
os.makedirs('reports', exist_ok=True)

# 1. SNP Dağılımı (rs429358 örneği)
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rs429358', hue='Disease', palette='viridis')
plt.title('Genotype Distribution vs Disease (rs429358)')
plt.savefig('reports/snp_distribution.png', dpi=300, bbox_inches='tight')
plt.close()

# 2. PRS Dağılımı
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='PRS', hue='Disease', fill=True, palette='magma')
plt.title('Polygenic Risk Score (PRS) Density Plot')
plt.savefig('reports/prs_distribution.png', dpi=300, bbox_inches='tight')
plt.close()

# 3. Manhattan Plot
plt.figure(figsize=(12, 6))
colors = ['#1f77b4' if p < 1.3 else '#ff7f0e' for p in analysis_df['-log10_P']]
plt.scatter(analysis_df['SNP'], analysis_df['-log10_P'], c=colors, s=100)
plt.axhline(y=1.3, color='r', linestyle='--')
plt.title('Genome-Wide Association Study (GWAS) - Manhattan Plot')
plt.ylabel('-log10(P-Value)')
plt.xticks(rotation=45)
plt.savefig('reports/manhattan_plot.png', dpi=300, bbox_inches='tight')
plt.close()

print("✅ Tüm görseller 'reports/' klasörüne kaydedildi!")
```
